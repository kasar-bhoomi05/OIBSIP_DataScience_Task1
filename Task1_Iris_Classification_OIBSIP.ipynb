# Step 1: Import Required Libraries

```python
# pandas is used for handling datasets in table form (DataFrames)
import pandas as pd

# matplotlib is used for plotting graphs and visualizations
import matplotlib.pyplot as plt

# sklearn provides built-in datasets and machine learning models
from sklearn.datasets import load_iris

# train_test_split helps divide data into training and testing sets
from sklearn.model_selection import train_test_split

# RandomForestClassifier is a powerful ML algorithm for classification
from sklearn.ensemble import RandomForestClassifier

# Metrics help us evaluate how good the model is performing
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

# Step 2: Load the Iris Dataset

```python
# load_iris() is a built-in dataset available in sklearn
# It contains information about iris flowers
iris = load_iris()

# Convert dataset into a pandas DataFrame (table format)
# iris.data contains the feature values (sepal/petal measurements)
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# iris.target contains the flower species encoded as numbers (0,1,2)
df["species"] = iris.target

# Convert numeric labels into actual species names for better readability
df["species"] = df["species"].map({
    0: "Setosa",
    1: "Versicolor",
    2: "Virginica"
})

# Display first 5 rows of dataset
print("Dataset Loaded Successfully!")
df.head()
```

# Step 3: Dataset Exploration (Understanding the Data)

```python
# Print total number of rows and columns
print("\nShape of Dataset (rows, columns):", df.shape)

# Check if there are any missing/null values
print("\nMissing Values in Dataset:\n", df.isnull().sum())

# Count how many samples belong to each species
print("\nNumber of flowers in each species:\n", df["species"].value_counts())
```

# Step 4: Data Visualization

```python
# Plot a bar chart to visualize distribution of species
df["species"].value_counts().plot(kind="bar")

# Add title and labels for clarity
plt.title("Distribution of Iris Flower Species")
plt.xlabel("Species")
plt.ylabel("Count")

# Show the graph
plt.show()
```

# Step 5: Split Dataset into Features and Target

```python
# Features (X) are input variables (flower measurements)
X = df.drop("species", axis=1)

# Target (y) is the output variable (species name)
y = df["species"]

# Split dataset into training and testing sets
# 80% data for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% test data
    random_state=42      # fixed value for reproducibility
)

print("\nData Split Completed Successfully!")
```

# Step 6: Train the Machine Learning Model

```python
# Random Forest Classifier:
# It builds multiple decision trees and combines them for better accuracy
model = RandomForestClassifier()

# Train the model using training data
model.fit(X_train, y_train)

print("\nModel Training Completed Successfully!")
```

# Step 7: Make Predictions


```python
# Predict species for test dataset
y_pred = model.predict(X_test)
```

# Step 8: Evaluate Model Accuracy


```python
# Accuracy tells us how many predictions were correct
accuracy = accuracy_score(y_test, y_pred)

print("\nModel Accuracy:", accuracy)
```

# Step 9: Detailed Classification Report


```python
# Classification report gives precision, recall, f1-score for each class
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

# Step 10: Confusion Matrix Visualization


```python
# Confusion Matrix shows correct vs incorrect predictions
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix in a visual format
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

# Add title for the plot
plt.title("Confusion Matrix - Iris Classification")

# Show matrix
plt.show()
```

# Step 11: Final Conclusion


```python
print("\nConclusion:")
print("The Random Forest Classifier achieved high accuracy in predicting Iris flower species.")
print("This project demonstrates a complete machine learning workflow:")
print("Data Loading  Visualization  Training  Testing  Evaluation")
```
